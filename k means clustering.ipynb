```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('Mall_Customers.csv')
df.head()
```

```python
data = df.iloc[:,[3,4]].values
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4)
```

```python
kmeans.fit(data)
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
unique, counts = np.unique(kmeans.labels_, return_counts=True)
```

```python
dict_X = dict(zip(unique, counts))
dict_X
```

```python
import seaborn as sns
```

```python
df["cluster"] = kmeans.labels_
```

```python
sns.lmplot('Annual Income (k$)', 'Spending Score (1-100)', data=df, hue='cluster', palette='coolwarm', size=6, aspect=1, fit_reg=False)
```

```python
kmeans.inertia_      #Inertia is the sum of squared error for each cluster. 
```

```python
kmeans.score
```
